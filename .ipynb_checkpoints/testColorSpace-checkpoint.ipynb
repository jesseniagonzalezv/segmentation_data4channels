```python
%matplotlib inline
%reload_ext autoreload
%autoreload 2

# Test color transforms in images
import numpy as np
import matplotlib.pyplot as plt
import cv2

img_path = 'dataset/data_512/images/0_0-0.npy'
img = np.load(img_path).transpose(1,2,0)

img2 = (img.astype(np.float32))/np.max(img)
print(np.shape(img2),np.min(img2),np.max(img2))

YUV = cv2.cvtColor(img2, cv2.COLOR_RGB2YUV)
YCbCr = cv2.cvtColor(img2, cv2.COLOR_RGB2YCrCb)
Lab = cv2.cvtColor(img2, cv2.COLOR_RGB2Lab)
Luv = cv2.cvtColor(img2, cv2.COLOR_RGB2Luv)
img3 = img2.astype(np.float32)+0.001 #to avoid division by 0
c1c2c3 = np.arctan(img3/np.dstack((cv2.max(img3[...,1], img3[...,2]), cv2.max(img3[...,0], img3[...,2]), cv2.max(img3[...,0], img3[...,1]))))

fig = plt.figure(figsize=(20, 20))  
#plt.subplots_adjust(hspace=0.2, wspace=0.2)
ax1 = plt.subplot(161)
ax1.imshow(img2), plt.title('RGB')
ax2 = plt.subplot(162)
ax2.imshow(YUV), plt.title('YUV')
ax3 = plt.subplot(163)
ax3.imshow(YCbCr), plt.title('YCbCr')
ax4 = plt.subplot(164)
ax4.imshow(Lab), plt.title('Lab')
ax5 = plt.subplot(165)
ax5.imshow(Luv), plt.title('Luv')
ax6 = plt.subplot(166)
ax6.imshow(c1c2c3), plt.title('C1C2C3')

```
